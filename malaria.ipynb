# Exploratory Data Analysis (EDA)

```python
print('\nTotal Number of Malaria Cases Reported in Africa:\n ', dataset['Malaria cases reported'].sum().astype(int))
```

```python
print('\nIncidence of malaria Reported in Africa:\n ', dataset['Incidence of malaria (per 1,000 population at risk)'].sum().astype(int))
```

```python
df1 = dataset.sort_values(by = "Malaria cases reported", ascending=False)
df1 = df1[df1["Year"]==2017][:20]
```

```python
#Top 20 countries with largest number of cases in 2007
fig1 = sns.catplot(x="Country Name", y="Malaria cases reported", kind="bar", data=df1, height=6, aspect=2)
plt.xticks(rotation=90, ha="right")
plt.title("Malaria Cases Against Country Name in 2007", fontsize=14, fontweight='bold')
plt.show()
fig1.savefig("fig1.png")
```

### This shows that in 2007, while Congo had the highest number of Malaria cases, Benin ranked 20th in lower to Congo

```python
df2 = dataset.sort_values(by = "Malaria cases reported", ascending=False)
df2 = df2[df2["Year"]==2007][:20]
```

```python
#Top 20 countries with largest number of cases in 2017
fig2 = sns.catplot(x="Country Name", y="Malaria cases reported", kind="bar", data=df2, height=6, aspect=2)
plt.xticks(rotation=90, ha="right")
plt.title("Malaria Cases Against Country Name in 2017", fontsize=14, fontweight='bold')
plt.show()
fig2.savefig("fig2.png")
```

### This shows that in 2017, Congo and Nigeria had taken a relative measure against malaria, thereby having Congo in the 5th position

```python
#Top 20 countries with largest number of Incidence of malaria (per 1,000 population at risk) in 2017
df3 = dataset[dataset["Year"]==2017][:20]
df3 = df3.sort_values(by = "Incidence of malaria (per 1,000 population at risk)", ascending=False)
```

```python
#Top 50 countries with largest number of deaths in 2017
fig3 = sns.catplot(x="Country Name", y="Incidence of malaria (per 1,000 population at risk)", kind="bar", data=df3, height=6, aspect=2)
plt.xticks(rotation=90, ha="right")
plt.title("Incidence of Malaria Against Country Name in 2007", fontsize=14, fontweight='bold')
plt.show()
fig3.savefig("fig3.png")
```

### This shows that in 2017, the cases of malaria was more acute in Burkina Faso, and less in Algeria

```python
#Relationship between number of cases and number of deaths reported
plt.figure(figsize =(16,8))
plt.title("Relationship between number of cases and Incidence of malaria (per 1,000 population at risk)", fontsize=14, fontweight='bold')
sns.scatterplot(x="Incidence of malaria (per 1,000 population at risk)", y="Malaria cases reported", data = dataset)
plt.savefig("fig4.png")
```

### This shows that in 2017, the severity of malaria was largely not dependent on the number of cases

```python
df4 = dataset.groupby("Year").sum().loc[:, ["Malaria cases reported"]]
```

```python
#Relationship between number of cases over the years
plt.figure(figsize =(16,8))
plt.title("Malaria cases reported over the years", fontsize=14, fontweight='bold')
sns.lineplot(data=df4)
plt.savefig("fig5.png")
```

### The line graph above shows the exponential increase in number of Malaria cases from year 2007 to 2017.

### The Plot shows a steady increase from 2007 to mid 2010 and the number of cases took a peak from mid 2010 to 2017, having heighest peak in 2017 with more than 100000000 reported cases in Africa alone. 

```python
fig = plt.figure(figsize =(16,8))
plt.subplots_adjust(left=None, bottom=None, right=None, top=1.5, wspace=None, hspace=None)
plt.subplot(211)
sns.lineplot(x="Year",y="Incidence of malaria (per 1,000 population at risk)", data=dataset.groupby("Year")["Incidence of malaria (per 1,000 population at risk)"].sum().reset_index(), color="red")
plt.title("Incidence of malaria (per 1,000 population at risk) over the years", fontsize=14, fontweight='bold')
plt.xticks(np.arange(2007, 2017, 1))
plt.savefig("fig6.png")
```

```python
fig = plt.figure(figsize =(16,8))
sns.barplot(x="Year", y="Incidence of malaria (per 1,000 population at risk)", data=dataset.groupby("Year")["Incidence of malaria (per 1,000 population at risk)"].sum().reset_index(), palette="coolwarm")
plt.title("Incidence of malaria (per 1,000 population at risk) over the years", fontsize=14, fontweight='bold')
plt.savefig("fig7.png")
```

### This shows that the severity of malaria has decreased over the years, thereby confirming that the severity does not depend on the number of cases

In Conclusion:

The analysis shows a steady increase in number of Malaria Cases and a stead decrease in its severity over the years.
